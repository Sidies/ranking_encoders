```python
import os
import sys
module_path = os.path.abspath(os.path.join('../..'))
if module_path not in sys.path:
    sys.path.append(module_path)
    
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../../../reports/best_scores.csv')
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(data)
```

```python
string_map = {
    'pointwise_normalized_regression': 'ptw. regression',
    'pointwise_classification': 'ptw. classification',
    'pointwise_ordinal_regression': 'ptw. ord. regression',
    'pairwise_classification': 'pw. classification',
    'listwise_regression': 'lstw. regression',
    'listwise_neural_network': 'lstw. neural network',
    'average_spearman': 'avg. spearman'
}

data = data.replace(string_map)

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(data)
```

```python
def get_data(data, approach, metric=['avg. spearman']):
    conditions = pd.Series(np.repeat(True, len(data)))
    conditions = conditions & (data['approach'] == np.full(len(data), approach)) if approach is not None else conditions
    conditions = conditions & (data['metric'] == np.full(len(data), metric)) if metric is not None else conditions
    return data[conditions]

def plot(data, name):
    # Filter data based on unique x or y values
    
    data = data.groupby('category').filter(lambda x: x['approach'].nunique() == len(data['approach'].value_counts()))
    sns.barplot(x=data['approach'], y=data['score'], hue=data['category'])
    plt.title('performance')
    plt.ylabel('avg. spearman')
    plt.xlabel('')
    if len(data['approach'].value_counts()) >= 3:
        plt.xticks(rotation=18)
    plt.legend(loc='lower left')
    plt.savefig(name + '_comparison.svg', bbox_inches='tight')
    plt.show()
    
def set_category(data, grouping, categories):
    def append_category(row):
        group = tuple(row[grouping].values[0])
        row['category'] = categories[group]
        return row
    grouped = data.groupby(grouping, dropna=False)
    return grouped.apply(append_category).reset_index(drop=True)
```

```python
approaches = [
    'regression',
    'ptw. regression',
    'ptw. classification',
    'ptw. ord. regression',
    'pw. classification',
    'lstw. regression',
    'lstw. neural network',
]

data_by_approach = {}
for approach in approaches:
    data_by_approach[approach] = get_data(data, approach, metric=['avg. spearman'])
    
for approach in approaches:
    data_by_approach[approach] = set_category(data_by_approach[approach], ['tuning'], {
        (np.nan,): 'default',
        ('bayes',): 'tuned',
        ('optuna',): 'tuned'
    })

data_categorized = pd.concat(data_by_approach.values())
```

```python
def plot_full(data, hue_order, name):
    sns.barplot(x=data['approach'], y=data['score'], hue=data['category'], hue_order=hue_order)
    plt.title('performance')
    plt.ylabel('avg. spearman')
    plt.xlabel('')
    if len(data['approach'].value_counts()) >= 3:
        plt.xticks(rotation=18)
    plt.legend(loc='lower left')
    plt.savefig(name + '.svg', bbox_inches='tight')
    plt.show()
```

```python
plot_full(pd.concat([
    data_by_approach[approach] for approach in [
        'regression',
        'ptw. regression',
        'ptw. classification',
        'ptw. ord. regression',
        'pw. classification',
        'lstw. regression',
        'lstw. neural network',
    ]
]), hue_order=['default', 'tuned'], name='performance_spearman')
```

```python
runtime_data = data[data['tuning'].isna() & data['metric'].isin(['avg. spearman'])]
runtime_data = runtime_data[['approach', 'score', 'runtime']]
runtime_data = runtime_data.groupby('approach').mean().reset_index()

display(runtime_data)
```

```python
runtime_data = runtime_data.set_index('approach').reindex([
    'regression',
    'ptw. regression',
    'ptw. classification',
    'ptw. ord. regression',
    'pw. classification',
    'lstw. regression',
    'lstw. neural network'
]).reset_index()

markers = ['o', 'o', '^', 'X', 'o', 'o', '^']
colors = ['red', 'orange', 'orange', 'orange', 'green', 'blue', 'blue']

sns.scatterplot(
    data=runtime_data, 
    x='runtime', 
    y='score', 
    hue='approach', 
    style='approach', 
    markers=markers, 
    palette=colors,
    s=150
)
plt.xlabel('runtime (s)')
plt.ylabel('avg. spearman')
plt.title('performance & runtime')
plt.legend(title='methods')
plt.savefig('runtimes.svg', bbox_inches='tight')
plt.show()
```
